```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/mobile_dataset.csv')
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df['price_range']
```

```python
from sklearn.feature_selection import SelectKBest, chi2

orf = SelectKBest(score_func=chi2, k=10)
```

```python
or_feature = orf.fit(x,y)
```

```python
dfscores = pd.DataFrame(or_feature.scores_, columns=['Score'])
dfcolumns = pd.DataFrame(x.columns)
```

```python
features_rank = pd.concat([dfcolumns,dfscores], axis=1)
```

```python
features_rank.columns = ['Features','Scores']
features_rank
```
