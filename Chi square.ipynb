```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('titanic')
```

```python
data.head()
```

```python
data.info()
```

```python
data = data[['sex', 'embarked', 'alone', 'pclass', 'survived']]
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

lbn = LabelEncoder()
```

```python
data['n_sex'] = lbn.fit_transform(data['sex'])
data['n_embarked'] = lbn.fit_transform(data['embarked'])
data['n_alone'] = lbn.fit_transform(data['alone'])
```

```python
data.head()
```

```python
data.drop(['sex','embarked','alone'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.rename(columns={'n_sex':'sex','n_embarked':'embarked','n_alone':'alone'})
```

```python
x = data.drop('survived', axis=1)
y = data['survived']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y , test_size=0.3, random_state=0)
```

```python
x_train.head()
```

```python
x_train.isnull().sum()
```

```python
#Performing Chi square test

from sklearn.feature_selection import chi2

chi2(x_train, y_train)
```

* Array 1 refers to F-value
    -- The most F-values are important
* Array 2 refers to P-value 
    -- The least P-values are important

```python
f_p_val = chi2(x_train, y_train)
p_val = pd.Series(f_p_val[1])
p_val.index = x_train.columns
p_val
```

### Observation

Sex column is the most important column when compared to the output output feature,Survived.
